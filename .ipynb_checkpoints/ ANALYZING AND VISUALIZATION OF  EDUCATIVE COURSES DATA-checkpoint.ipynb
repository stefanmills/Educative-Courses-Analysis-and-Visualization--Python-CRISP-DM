```python
import os
import zipfile
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
zip_path="C:\\Users\\user\\Desktop\\python project\\dataset.zip"
ext_path="C:\\Users\\user\\Desktop\\python project"

#creating the extracted folder
if not os.path.exists(ext_path):
    os.makedirs(ext_path)
#Unziping file

with zipfile.ZipFile(zip_path,'r') as zipping:
    zipping.extractall(ext_path)
```

```python
#Merging Files
all_files=glob.glob("C:\\Users\\user\\Desktop\\python project\\Data Sheet Educative Courses Export\\Data_Sheet*.csv")
all_files

df_list=[pd.read_csv(file) for file in all_files] #Created a list of dataframes form the all_files list
all_csvs=pd.concat(df_list,ignore_index=True)
all_csvs.to_csv('merged_data.csv')
all_csvs
```
